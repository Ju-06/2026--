## fillna(Value)

```python
import pandas as pd
import numpy as np

df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3,4,np.nan,1],
                   [np.nan, np.nan, np.nan, np.nan],
                  [np.nan, 3, np.nan, 4]],
                  columns=list("ABCD"))
df
```

```python
df.fillna(10)
# nan값에 전부 10을 대입
```

```python
df
# fillna 를 써도 기존 데이터 프레임은 변경xx
```

```python
df.fillna('missing', axis=0), df.fillna('missing')
```

```python
df.fillna('missing', limit=1, axis=1)
# 세로로 내려가면서 limit = n, 최대 n만 바꾸기
# df.fillna('missing', limit=1, axis=1) axis가 0일때는 세로 방향으로 조사, axis가 1이면 가로 방향으로 조사
```

```python
s = pd.Series([100,200,400],index=list("ACD"))
s
```

```python
from IPython.display import display_html
def display_side_by_side(*args):
    """여러 데이터프레임 비교가 쉽게 옆쪽으로 표시한다"""
    html_str=''
    for df in args:
        html_str += df.to_html() + '&nbsp;'*4
    display_html(html_str.replace('table','table style="display:inline"'), raw=True)
```

```python
display_side_by_side(df, df.fillna("missing", limit=1, axis=1))
```

```python
df2 = pd.DataFrame(-np.ones((4,4)), columns=list('ABCE'))
df2
```

```python
display_side_by_side(df, df.fillna(df2))
```
