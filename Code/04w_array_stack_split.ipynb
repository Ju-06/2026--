## numpy vstack()

```python
import numpy as np

a = np.arange(6).reshape(3,2)
a
```

```python
b = np.arange(6,12).reshape(3,2)
b
```

```python
np.vstack((a,b))
```

```python
c = np.arange(20,24).reshape(2,2)
c
```

```python
np.vstack((a,c))
```

```python
np.vstack((a,b,c))
```

```python
d = np.array([100, 200])
np.vstack((a,d))
```

```python
d = np.array([100, 200, 300]) #(2,2)배열이랑 1차원 배열이라 오류
np.vstack((a,d))
```

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
np.vstack((x,y))
```
