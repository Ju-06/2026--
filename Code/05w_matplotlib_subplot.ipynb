```python
import matplotlib.pyplot as plt

# 데이터 생성
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# 선 플롯 생성
plt.plot(x, y, marker='o', linestyle='--', color='b', label='소수(Prime Numbers)')

# 축 및 제목 설정
plt.xlabel('X-축(axis)')
plt.ylabel('Y-축(axis)')
plt.title('소수 그리기(Prime Numbers Plot)')

# 범례 표시
plt.legend()

# 플롯 표시
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 난수 고정 (재현 가능성)
np.random.seed(2025)

N = 80
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)

# 점 크기 설정 (0~15 범위 느낌)
area = (20 * np.random.rand(N)) ** 2

# 산점도 생성
plt.scatter(x, y, s=area, c=colors, alpha=0.7)

# 컬러바 표시
plt.colorbar()

# 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000) # 1000개의 랜덤 값 추출
y = np.random.randint(-100, 100, 1000)

mask1 = abs(x) > 50
mask2 = abs(y) > 50

x = x[mask1 + mask2]
y = y[mask1 + mask2]

plt.scatter(x, y, c = x, cmap='jet', alpha= 0.4)
plt.colorbar()
plt.show()
```

## 도수분포표

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

plt.hist([1,1,1,2,2,3,4,5,6,6,6,6], rwidth= .98, bins=[.5 ,1.5 ,2.5 ,3.5,4.5,5.5,6.5])
# bins는 구간의 개수 또는 구간의 경계를 지정
plt.xticks(range(1,7))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. 1부터 6까지의 숫자를 정해진 확률(p)에 따라 1,000,000번 추출합니다.
# p의 합은 1이 되어야 하며, 여기서는 3이 나올 확률이 0.3으로 가장 높습니다.
dice = np.random.choice(range(1, 7), 1000000,
                        p=[0.15, 0.25, 0.3, 0.1, 0.1, 0.1])

# 2. 그래프 크기와 해상도 설정
plt.figure(figsize=(5, 3), dpi=200)

# 3. 히스토그램 그리기
# bins=np.arange(0.5, 7)은 막대가 숫자 1~6의 중앙에 오도록 범위를 설정합니다.
plt.hist(dice, rwidth=0.98, bins=np.arange(0.5, 7))

# 4. x축 눈금을 1부터 6까지 표시
plt.xticks(range(1, 7))

# 5. 결과 저장 및 출력
plt.savefig('주사위.png')
plt.show()
```

# plt.subplot()

```python
import matplotlib.pyplot as plt

# 전체 제목 - 어느 위치에서 호출해도 무관 (Figure 레벨)
plt.suptitle('전체 제목', fontsize=14, y=1.02)

# subplot 1 - 생성 직후 바로 title 지정
plt.subplot(2, 3, 1)
plt.plot([1, 4, 9, 16])
plt.title('제곱수')

# subplot 5
plt.subplot(2, 3, 6)
plt.plot([1, 2, 3, 4])
plt.title('선형')

# subplot 2
plt.subplot(2, 3, 2)
plt.plot([6, 5, 2, 1])
plt.title('감소')

plt.tight_layout()
plt.show()
```

# plt.subplots() 그림 제목

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 3, figsize=(7, 4))
fig.set_size_inches(6, 3)

ax[0, 1].plot([1, 4, 9, 16])
ax[1, 0].plot([1, 2, 3, 4])
ax[1, 2].plot([6, 5, 2, 1])

# 전체 제목 — Figure 레벨
fig.suptitle('전체 제목', fontsize=14, fontweight='bold')

# 각 subplot 제목 — Axes 레벨
ax[0, 0].set_title('(0,0) 빈 플롯')
ax[0, 1].set_title('(0,1) 제곱수')
ax[0, 2].set_title('(0,2) 빈 플롯')
ax[1, 0].set_title('(1,0) 선형')
ax[1, 1].set_title('(1,1) 빈 플롯')
ax[1, 2].set_title('(1,2) 감소')

plt.tight_layout() # 적당히 알아서 비율 맞춰주는 코
plt.show()
```

# fig.add_subplot() 그림 제목

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(6, 4))

ax1 = fig.add_subplot(231)
ax1.plot([1, 4, 9, 16])
ax1.set_title('제곱수')

ax2 = fig.add_subplot(233)
ax2.plot([1, 2, 3, 4])
ax2.set_title('선형')

ax3 = fig.add_subplot(235)
ax3.plot([6, 5, 2, 1])
ax3.set_title('감소')

fig.suptitle('전체 제목', fontsize=14)
plt.tight_layout()
plt.show()
```
